# Ler e percorrer uma imagens colorida


```python
from PIL import Image
import numpy as np

img = Image.open("/content/lena_color.png")
img = np.array(img)

w, h, nc = img.shape

for x in range(w):
  for y in range(h):
    for c in range(nc):
      print(img[x,y,c])
```

Exercício 01

Ler uma imagem colorida e converter usando os três métodos: </br>

Média (R + G + B) / 3 </br>
Brilho (max(R, G, B) + min (R, G, B)) / 2 </br>
Luminosidade 0.21 R + 0.71 G + 0.07 B

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Carregando a imagem colorida
img = Image.open("/content/lena_color.png")
image_color = np.array(img)

print("Imagem colorida original")
plt.imshow(image_color)
plt.show()
```

## Média: (R + G + B) / 3

```python
## Converte a matriz para float para evitar o erro de 'overflow' na soma
image_color_float = image_color.astype(np.float64)

# Calculando a média
image_gray_mean = (image_color_float[:, :, 0] + image_color_float[:, :, 1] + image_color_float[:, :, 2]) / 3

# Converte o resultado de volta para o tipo uint8 (0-255)
image_gray_mean = image_gray_mean.astype(np.uint8)

print("Imagem em escala de cinza com média")
plt.imshow(image_gray_mean, cmap='gray')
plt.show()
```

# Brilho: (max(R, G, B) + min (R, G, B)) / 2

```python
# Cálculo do brilho
brilho_max = np.max(image_color, axis=2)
brilho_min = np.min(image_color, axis=2)
brilho = (brilho_max + brilho_min) / 2

scale_image_gray = brilho.astype(np.uint8)

print("\n\nImagem em escala de cinza com brilho")
plt.imshow(scale_image_gray, cmap='gray')
plt.show()
```

Luminosidade: 0.21 R + 0.71 G + 0.07 B

```python
# Separando os canais de cor
R = image_color_float[:, :, 0]
G = image_color_float[:, :, 1]
B = image_color_float[:, :, 2]

# Aplicando a fórmula de luminosidade
image_gray_luminosity = 0.21 * R + 0.71 * G + 0.07 * B

# Convertendo o resultado para o tipo de imagem (uint8)
image_gray_luminosity = image_gray_luminosity.astype(np.uint8)

# Exibindo a imagem em escala de cinza
print("Imagem em escala de cinza com base na luminosidade")
plt.imshow(image_gray_luminosity, cmap='gray')
plt.show()
```
